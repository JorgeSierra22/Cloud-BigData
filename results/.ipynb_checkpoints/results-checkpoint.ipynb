```python
import pandas as pd
import matplotlib.pyplot as plt
```

Los primeros datos que queremos analizar son las horas en las que se produjo el mayor numero de crimes 

```python
df = pd.read_csv("hours.csv", header=None, names=['Hora', 'Cantidad'])
```

```python
df["Cantidad"].plot(kind="bar")

plt.show()
```

```python
media_ponderada = (df['Hora'] * df['Cantidad']).sum() / df['Cantidad'].sum()
media_ponderada
```

```python
desviacion_estandar = df['Hora'].std()
desviacion_estandar
```

```python
estadisticas = df['Cantidad'].describe()
print(estadisticas)
```

```python
df.plot(x='Hora', y='Cantidad', marker='o')
plt.title('Tendencia de Crmenes por Hora del Da')
plt.xlabel('Hora del Da')
plt.ylabel('Cantidad')
plt.show()
```

```python
df.boxplot(column='Cantidad', by='Hora')
plt.title('Boxplot de Crmenes por Hora del Da')
plt.suptitle('')  # Elimina el ttulo automtico
plt.xlabel('Hora del Da')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Calcular la media
media = df['Cantidad'].mean()

# Crear la grfica de barras
plt.bar(df['Hora'], df['Cantidad'], color='blue', alpha=0.7, label='Cantidad')
plt.axhline(y=media, color='red', linestyle='--', label='Media')

# Configurar el ttulo y las etiquetas
plt.title('Cantidad por Hora')
plt.xlabel('Hora')
plt.ylabel('Cantidad')

# Aadir una leyenda
plt.legend()

# Mostrar la grfica
plt.show()
```

**Conclusiones:**
    
  Primero, con respecto a los resultados obtenidos sobre cuales son las horas en las que se producen mayores crmenes, gracias a estas grficas, podemos ver que los delitos se concentran en la franja horaria comprendida entre las 18 y las 21 horas, a pesar de no ser el pico en cuanto a nmero (este se produce a las 12 de la noche). Esto nos hace ver que la mayora de los crmenes en Chicago se producen en horas en las cuales la mayora de la poblacin est regresando de sus trabajos, y por lo tanto, hay una mayor actividad tanto en las calles, como en los restaurantes y el transporte, haciendo de estas 4 horas la oportunidad perfecta para llevar a cabo actividades delictivas.

```python
df3 = pd.read_csv("typeOfCrime.csv", header=None, names=['Type',"Year", 'Cantidad'])
```

```python
df3
```

```python
# Agrupar por la columna 'Type' y sumar las cantidades
sum_by_type = df3.groupby('Type')['Cantidad'].sum().reset_index()

# Imprimir el resultado
print(sum_by_type)
```

```python
suma = sum_by_type["Cantidad"].sum()
print(suma)
```

```python
sum_by_type.sort_values(by='Cantidad', ascending=False).head(5)
```

```python
sum_by_type.plot(x='Type', y='Cantidad',kind="bar")

plt.show()
```

```python
df_sorted_Year = df3.sort_values(by='Year', ascending=True)
df_sorted_Year
```

```python
#Crmenes ms repetidos en 2023
df_filtered = df3[df3['Year'] == 2023]
df_2023 = df_filtered.sort_values(by='Cantidad', ascending=False)
print(df_2023.head())
```

```python
#Crmenes ms repetidos que sobrepasan una cantidad preocupante 
df_most_repeated_crimes = df3[df3['Cantidad'] >= 65000]
print(df_most_repeated_crimes.sort_values(by='Year', ascending=True))
```

```python
# Contar el nmero de veces que cada ao est presente
year_counts = df_most_repeated_crimes['Year'].value_counts().sort_index()
print(year_counts)
```

```python
# Graficar el resultado
year_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Year')
plt.ylabel('Nmero de veces')
plt.title('Nmero de veces que cada Ao est presente')
plt.show()
```

```python
top_crimes_by_year = pd.DataFrame(columns=df3.columns)

# Obtener la lista de aos nicos en tu DataFrame
unique_years = df3['Year'].unique()

# Iterar sobre cada ao y obtener los 5 crmenes ms repetidos
for year in unique_years:
    # Filtrar el DataFrame para el ao actual
    df_filtered = df3[df3['Year'] == year]
    
    # Ordenar por 'Cantidad' de forma descendente
    df_year_top5 = df_filtered.sort_values(by='Cantidad', ascending=False).head(5)
    
    # Concatenar los resultados al DataFrame principal
    top_crimes_by_year = pd.concat([top_crimes_by_year, df_year_top5])

# Mostrar el DataFrame final con los 5 crmenes ms repetidos por ao
print(top_crimes_by_year)
```

```python
import seaborn as sns 
sns.set(style="whitegrid")

# Configurar el tamao del grfico
plt.figure(figsize=(12, 8))

# Crear un grfico de barras agrupado por ao y tipo de crimen
sns.barplot(x='Type', y='Cantidad', hue='Year', data=top_crimes_by_year)

# Configurar etiquetas y ttulo del grfico
plt.xlabel('Tipo de Crimen')
plt.ylabel('Cantidad')
plt.title('Top 5 Crmenes Ms Repetidos por Ao')

# Ajustar la leyenda
plt.legend(title='Ao', bbox_to_anchor=(1, 1))

# Rotar las etiquetas del eje x para mayor claridad
plt.xticks(rotation=45, ha='right')

# Mostrar el grfico
plt.show()
```

```python
top_crimes_by_year2001=top_crimes_by_year[top_crimes_by_year['Year'] == 2001]
print(top_crimes_by_year2001)
```

```python
top_crimes_theft=top_crimes_by_year[top_crimes_by_year['Type'] == "THEFT"]
print(top_crimes_theft.sort_values(by='Cantidad', ascending=False))
```

```python
top_crimes_theft.sort_values(by='Year', ascending=True).plot(x='Year', y='Cantidad', marker='o')

plt.show()
```

```python
top_crimes_Battery=top_crimes_by_year[top_crimes_by_year['Type'] == "BATTERY"]
print(top_crimes_Battery.sort_values(by='Cantidad', ascending=False))
```

```python
top_crimes_Battery.sort_values(by='Year', ascending=True).plot(x='Year', y='Cantidad', marker='o')

plt.show()
```
